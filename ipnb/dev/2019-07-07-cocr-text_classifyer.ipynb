```python
TEXT = "../../src/spekuloom/Textos Alunos.txt"
OUTPUT_DIR = "/home/carlo/Documentos/dev/spekuloom/src/SpekuloomS/basico/"
def mangle_name(nomes):
    nome = (nomes[0][1:8].lower() + "_"+ "_".join(a[:4//len(nomes)].lower() for a in nomes[1:3])).ljust(12) + nomes[0][0]
    nome = nome
    nome = nome.translate(str.maketrans(" .áãâéêíóôõú", "__aaaeeiooou"))
    return nome
with open(TEXT, "r") as textfile:
    text = textfile.read()
    texts = text.split("@")[1:]
    only_texts = ["\n".join(t.split("\n")[1:]) for t in texts]
    only_texts_heads = [mangle_name(t.split("\n")[0].split())
                        for t in texts]
    heads_and_texts = zip(only_texts_heads, only_texts)
    #[print(h, "\t", len(t)) for h, t in heads_and_texts]

    for head, out_text in heads_and_texts:
        filename = OUTPUT_DIR+f"{head}.txt"
        if len(out_text) <700:
            continue
        with open(filename, "w") as out_file:
            print(filename)
            out_file.write(out_text)
    '''    '''
```

```python
import csv
with open(SPEKULOOM_PATH+"spekuloom/learn.tab", 'r') as csvfile:
    learn = csv.reader(csvfile, delimiter='\t', quotechar='|')
    h, t, c = next(learn), next(learn), next(learn)
    with open(SPEKULOOM_PATH+"spekuloom/samples.tab", 'r') as csvsample:
        sample = csv.reader(csvsample, delimiter='\t', quotechar='|')
        head = next(sample)
        ts, cs = next(sample), next(sample)
        #print(line)
        cols = [h.index(col) for col in head if col in h]
        body = [h, t, c]+[[0 if lin_index not in cols else line[cols.index(lin_index)] for lin_index in range(len(h))]
                for line in sample]
        print(cols, len(cols))
        print(body)
        with open(SPEKULOOM_PATH+"spekuloom/samples.tab", 'w') as csvout:
            cw = csv.writer(csvout, delimiter='\t', quotechar='|')
            [cw.writerow(row) for row in body]
```

```python
import csv
with open("learn.tab", 'r') as csvfile:
    grammat = csv.reader(csvfile, delimiter='\t', quotechar='|')
    grammat = [l for l in grammat]
    grammat, labels = [[int(g) for g in l[1:-1]] for l in grammat[3:]], grammat[0][1:-1]
    print(labels)
    #print(grammat)
    xgram = zip(* grammat)
    wgram = [(label, sum(gram[0:11]), sum(gram[11:22]), sum(gram[22:])) for label, gram in zip(labels, xgram)]
    print(wgram)
    with open("gram.tab", 'w') as csvout:
        fieldnames = ['patt', 'ba', 'it', 'tr']

        cw = csv.writer(csvout, delimiter='\t', quotechar='|')
        [cw.writerow(row) for row in [fieldnames]+wgram]
```
