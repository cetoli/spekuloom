```python
TEXT = "../../src/spekuloom/Textos Alunos.txt"
OUTPUT_DIR = "/home/carlo/Documentos/dev/spekuloom/src/SpekuloomS/todos/"
with open(TEXT, "r") as textfile:
    text = textfile.read()
    texts = text.split("@")[1:]
    only_texts = ["\n".join(t.split("\n")[1:]) for t in texts]
    [print(len(t)) for t in only_texts]
    for index, out_text in enumerate(only_texts):
        filename = OUTPUT_DIR+f"sample_{index}.txt"
        if len(out_text) <700:
            continue
        with open(filename, "w") as out_file:
            print(filename)
            out_file.write(out_text)
```

```python
import csv
with open("learn.tab", 'r') as csvfile:
    grammat = csv.reader(csvfile, delimiter='\t', quotechar='|')
    grammat = [l for l in grammat]
    grammat, labels = [[int(g) for g in l[1:-1]] for l in grammat[3:]], grammat[0][1:-1]
    print(labels)
    #print(grammat)
    xgram = zip(* grammat)
    wgram = [(label, sum(gram[0:11]), sum(gram[11:22]), sum(gram[22:])) for label, gram in zip(labels, xgram)]
    print(wgram)
    with open("gram.tab", 'w') as csvout:
        fieldnames = ['patt', 'ba', 'in', 'tr']

        cw = csv.writer(csvout, delimiter='\t', quotechar='|')
        [cw.writerow(row) for row in [fieldnames]+wgram]
```
